```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dnd_ability_score_data.csv')
```

```python
sns.pairplot(df, hue="Main Class")
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
linkage = AgglomerativeClustering(n_clusters=5)
```

```python
simple_data = np.array(df.drop(['Main Class', 'Race'], axis=1))
print(simple_data)
```

```python
linkage.fit(simple_data)
linkage.labels_
```

```python
df['Linkage'] = linkage.labels_
df
```

```python
sns.pairplot(df, hue="Linkage")
```
