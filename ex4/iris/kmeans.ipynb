```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
```

```python
sns.pairplot(df, hue="variety")
```

### Verificar, através do Elbow method, o número otimizado de grupos

```python
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:, [0, 1, 2, 3]]

squared_errors = []

k_range = np.arange(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(x)
    squared_errors.append(km.inertia_)
```

```python
plt.plot(k_range, squared_errors)
plt.xlabel('K value')
plt.ylabel('Sum of Squared Error')
plt.title('Elbow Curve')
```

### Algoritmo de KMeans utilizando 3 como K (por conta do Elbow)

```python
kmeans = KMeans(n_clusters=3)
```

```python
simple_data = np.array(df.drop('variety', axis=1))
print(simple_data)
```

```python
kmeans.fit(simple_data)
kmeans.labels_
```
